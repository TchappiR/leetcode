## Exercice 1

Ecris une fonction qui prend deux mots en entrée et compte les occurences de chaque lettre afin de dire si les nomts sont des anagrammes

```python
def anagramme(word1 : str, word2 : str) -> bool :

    if len(word1) != len(word2):
        return False

    counter_word1 , counter_word2 = {}, {}

    for index in range(len(word1)):
        counter_word1[word1[index]] = 1 + counter_word1.get(word1[index], 0)
        counter_word2[word2[index]] = 1 + counter_word2.get(word2[index], 0)
    return counter_word1 == counter_word2


if __name__ == "__main__":
    print(anagramme("banane", "anane"))
```

## Exercice 2

Ecrire une fonction qui prend en entrée une liste de nombre et une valeur cible et renvoie en sortie, une liste cotenant les indexs de deux elements de la liste dont la somme est la valeur cible.

```python
def twoSum(nums: list[int], target: int) -> list[int]:
    answer = []
    for index1, num in enumerate(nums):
        for index2, other_num in enumerate(nums[index1+1:]):
            sum = nums[index1] + nums[index1+index2+1]
            if sum == target :
                answer = [index1, index1+index2+1]
                return answer
```

## Exercice 3

Crée une fonctions qui prends une liste d'elements et renvoie la souliste de tous les anagrammes de la liste proncipale

```python
def anagramme(word : str, other_word : str) -> bool:
    if sorted(list(word.lower())) == sorted(list(other_word.lower())):
        return True
    return False
```

```python
def group_anagramme(strs : list[str]) -> list[list[str]]:
    indexs_use = []

    dic_anagram = dict()
    cpt = 0

    for pos, elt in enumerate(strs):
        if pos not in indexs_use:
            dic_anagram[cpt] = [elt]
            indexs_use.append(pos)
            for cursor, wrd in enumerate(strs[pos+1:]):
                pos2 = pos + cursor + 1
                if pos2 not in indexs_use:
                    if anagramme(elt, wrd):
                        dic_anagram[cpt].append(wrd)
                        indexs_use.append(pos2)
        cpt += 1
    return list(dic_anagram.values())
```

```python
group_anagramme(["act","pots","tops","cat","stop","hat"])
```

```python
group_anagramme(strs = ["x"])
```
