## Exercice 1

Ecris une fonction qui prend deux mots en entrÃ©e et compte les occurences de chaque lettre afin de dire si les nomts sont des anagrammes

```python
def anagramme(word1 : str, word2 : str) -> bool :

    if len(word1) != len(word2):
        return False

    counter_word1 , counter_word2 = {}, {}

    for index in range(len(word1)):
        counter_word1[word1[index]] = 1 + counter_word1.get(word1[index], 0)
        counter_word2[word2[index]] = 1 + counter_word2.get(word2[index], 0)
    return counter_word1 == counter_word2


if __name__ == "__main__":
    print(anagramme("banane", "anane"))
```
